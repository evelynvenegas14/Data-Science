Luego definimos nuestras funciones de media y desviación estándar

```python
def media(datos):
  n=len(datos)
  suma=0
  for i in range(n):
    suma=suma+datos[i]
  return suma/n

def desviacion_estandar(datos):
  n=len(datos)
  suma=0
  for i in range(n):
    suma=suma+(datos[i]-media(datos))**2
  return (suma/n)**0.5
```

Ahora, haremos una función para generar la curva de probabilidad

```python
import numpy as np
def curva_probabilidad(datos):
 u=media(datos)
 s=desviacion_estandar(datos)

 x=np.linspace(u-5*s,u+5*s,100) #guardara los datos de la curva de probabilidad
 return x
```

Antes de usarla, guardamos todas las medias y desviaciones estándar en arrays para poder usarlas después al calcular las curvas de probabilidad

```python
setosa=[longitudes_sepalo_setosa,ancho_sepalo_setosa,longitud_petalo_setosa,ancho_petalo_setosa] #hacemos conjuntos de listas para reducir el numero de lineas
versicolor=[longitudes_sepalo_versicolor,ancho_sepalo_versicolor,longitud_petalo_versicolor,ancho_petalo_versicolor]
virginica=[longitudes_sepalo_virginica,ancho_sepalo_virginica,longitud_petalo_virginica,ancho_petalo_virginica]

#medias de la setosa
u_setosa = []
for i in range(len(setosa)):
  u_setosa.append(media(setosa[i]))
  print(u_setosa)

#desviaciones estandar de la setosa
s_setosa = []
for i in range(len(setosa)):
  s_setosa.append(desviacion_estandar(setosa[i]))
  print(s_setosa)

#medias de la versicolor
u_vscolor = []
for i in range(len(versicolor)):
  u_vscolor.append(media(versicolor[i]))
  print(u_vscolor)

#desviaciones estandar de la versicolor
s_vscolor = []
for i in range(len(versicolor)):
  s_vscolor.append(desviacion_estandar(versicolor[i]))
  print(s_vscolor)

#medias de la virgnica
u_virginica = []
for i in range(len(virginica)):
  u_virginica.append(media(virginica[i]))
  print(u_virginica)

#desviaciones estandar de la virginica
s_virginica = []
for i in range(len(virginica)):
  s_virginica.append(desviacion_estandar(virginica[i]))
  print(s_virginica)
```

Ahora, usamos la función de curva de probabilidad para calcular todas estas curvas

```python
x_setosa_curvas = []
for i in range(len(setosa)):
    x_setosa = curva_probabilidad(setosa[i])
    x_setosa_curvas.append(x_setosa)

x_versicolor_curvas = []
for i in range(len(versicolor)):
    x_versicolor = curva_probabilidad(versicolor[i])
    x_versicolor_curvas.append(x_versicolor)

x_virginica_curvas = []
for i in range(len(virginica)):
    x_virginica = curva_probabilidad(virginica[i])
    x_virginica_curvas.append(x_virginica)

for x_setosa in x_setosa_curvas: #solo para asegurarnos que si hace lo que debe
  print(x_setosa)
for x_versicolor in x_versicolor_curvas:
  print(x_versicolor)
for x_virginica in x_virginica_curvas:
  print(x_virginica)
```
